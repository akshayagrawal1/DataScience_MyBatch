def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray

```python
img = cv2.imread('260px-Virat..Kohli.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
print(img.shape)
print(gray.shape)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(gray, 'gray')
plt.show()
```

## Convolutional of the image

```python
from scipy import signal
```

```python
k = np.array([[-1,-1,-1],
              [-1,8,-1],
              [-1,-1,-1]])

print(k)

k1 = np.array([[1,2,1],
              [2,4,2],
              [1,2,1]])/16.0

print(k1)
```

```python
plt.figure(figsize=(6,6))
conv = signal.convolve2d(gray,k1)
plt.imshow(conv,"gray")
```

```python
plt.figure(figsize=(6,6))
relu = np.maximum(conv,0  )
plt.imshow(relu,'gray')
```

# Pooling

```python
cut1 = np.arange(0, img.shape[0], 2)
y1 = np.maximum.reduceat(relu, cut1)
cut2 = np.arange(0, img.shape[1], 2)
pool = np.maximum.reduceat(y1, cut2,axis=1)
plt.figure(figsize=(6,6))
plt.imshow(pool,'gray')
```
