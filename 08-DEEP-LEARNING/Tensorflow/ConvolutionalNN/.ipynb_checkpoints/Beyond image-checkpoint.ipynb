```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from matplotlib import image
import cv2
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
img = image.imread('image4.jpg')
gray = rgb2gray(img)
```

```python
print(img.shape)
print(gray.shape)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(gray, 'gray')
plt.show()
```

```python
plt.figure(figsize=(15,15))
conv = signal.convolve2d(gray,k)
plt.imshow(conv,"gray")
```

```python
plt.figure(figsize=(15,15))
relu = np.maximum(conv,0  )
plt.imshow(relu,'gray')
```

```python
cut1 = np.arange(0, img.shape[0], 2)
y1 = np.maximum.reduceat(relu, cut1)
cut2 = np.arange(0, img.shape[1], 2)
pool = np.maximum.reduceat(y1, cut2,axis=1)
plt.figure(figsize=(15,15))
plt.imshow(pool,'gray')
```
