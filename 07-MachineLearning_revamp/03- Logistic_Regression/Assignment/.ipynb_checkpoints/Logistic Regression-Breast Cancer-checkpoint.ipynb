```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**Step-1: Read the data set**

    1. Sample code number: id number 
    2. Clump Thickness: 1 - 10 
    3. Uniformity of Cell Size: 1 - 10 
    4. Uniformity of Cell Shape: 1 - 10 
    5. Marginal Adhesion: 1 - 10 
    6. Single Epithelial Cell Size: 1 - 10 
    7. Bare Nuclei: 1 - 10 
    8. Bland Chromatin: 1 - 10 
    9. Normal Nucleoli: 1 - 10 
    10. Mitoses: 1 - 10 
    11. Class: (2 for benign, 4 for malignant)


```python
columns = ['idnumber','clumpthick','unicellsize','unicellshape','adhesion','epithelcellsize',
          'barenuclei','chromatin','normnucle','mitoses','class']
```

```python
data = pd.read_csv('breast-cancer-wisconsin.data')
data.columns = columns
data.head()
```

```python
data.drop('idnumber',axis=1,inplace=True)
data.head()
```

```python
corr = data.corr()
sns.heatmap(corr,annot=True)
```

```python
data.keys()
```

```python
# Clump Thickness,  Uniformity of Cell Size
df = data[['clumpthick','unicellshape','class']]
df['class'].replace(to_replace=[2,4],value=[0,1],inplace=True)
```

```python
sns.relplot(x='clumpthick',y='unicellshape',data=df,hue='class')
```

```python
fig = sns.FacetGrid(data =df, hue='class',aspect=4)
fig.map(sns.kdeplot,'clumpthick',shade=True)
```

```python
fig = sns.FacetGrid(data =df, hue='class',aspect=4)
fig.map(sns.kdeplot,'unicellshape',shade=True)
```
