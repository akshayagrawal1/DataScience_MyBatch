# Linear Regression


> ### $\hat y = a + b*X$
    b = coefficient
    a = intercept

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

# working on a car dataset which contains
- speed(mph)
- stopping distance (ft)

```python
sns.set(color_codes=True)
```

```python
cars = pd.read_csv('../dataset/cars.csv')
cars.head()
```

```python
cars.info()
```

### Discriptive Stats

```python
sns.FacetGrid(data=cars, aspect = 4)
sns.kdeplot(data=cars['speed'],shade=True)
sns.kdeplot(data=cars['dist'],shade=True)
plt.show()
sns.FacetGrid(data=cars, aspect = 4)
plt.subplot(1,2,1)
sns.distplot(a=cars['speed'])
plt.subplot(1,2,2)
sns.distplot(a=cars['dist'])
plt.show()
```

```python
cars.keys()
```

```python
#join plot
sns.jointplot('speed','dist',data=cars,kind='scatter')
plt.show()
sns.jointplot('speed','dist',data=cars,kind='kde')
plt.show()
```

```python
import statsmodels.formula.api as smf
```

```python
cars.keys()
```

```python
model = smf.ols(formula="dist~speed",data=cars)
results = model.fit()
```

# Covariance

```python
cars.cov()
```

>### Correlation coefficient (*r* )

```python
cars.corr()
```

```python
print(results.summary())
```

```python
results.params
```

```python
results.rsquared
```

```python
import statsmodels.api as sm
```

```python
fig = plt.figure(figsize=(12,8))
fig=sm.graphics.plot_regress_exog(results,'speed',fig=fig)
```

```python
sns.lmplot(x='speed',y='dist',data=cars,aspect=3)
```
