```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load data
nyb = pd.read_table('./data/nybirths.dat',header=None) # reading table data
nyb.columns = ['births']
nyb_array = nyb.values # converting data into array
# Creating DataFrame (14 rows, and 12 columns)
start = 1946 # year of starting
years = len(nyb) / 12 # calculating number of year in data 
columns= ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'] # Months
index = np.arange(start,start+years,dtype=int) #  years in array 
try:
    
    df = pd.DataFrame(nyb_array.reshape((int(years),12)),index = index, columns=columns) # creating DataFrame
except:
    print('INFO: Cannot build array ({},12)'.format(years))

df
```

```python
nyb = nyb.reset_index()
```

```python
plt.subplots(figsize=(15,4))
sns.lineplot(x='index',y='births',data=nyb)
```

```python
sns.catplot(x='births',data=nyb,kind='box')
```
