```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="darkgrid",color_codes=True)
```

## **Loading preprocessed data-set**

```python
inocar = pd.read_excel('inocar.xlsx')
inocar.head()
```

```python
inocar.dropna(axis=0,inplace=True)
```

```python
sns.catplot(x="am", y="kpl", data=inocar)
```

```python
inocar.am.unique()
```

```python
sns.catplot(x="am", y="kpl", kind = 'swarm',data=inocar)
```

```python
sns.catplot(x="et", y="kpl", kind = 'boxen',data=inocar, height = 6, aspect=2)
plt.show()
```
