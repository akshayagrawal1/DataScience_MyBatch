```python
# Number 
# Read and convert into gray scale
num = cv2.imread('number5.jpg') # reading in BGR
# Step-1 : convert into grayscale
gray = cv2.cvtColor(num,cv2.COLOR_BGR2GRAY)
# Step-2 : resize into 8 x 8
gray_re = cv2.resize(gray,(8,8))
# Step-3: NOT operation
gray_not = cv2.bitwise_not(gray_re)
# Step-4: reshape in 1 x 64
gray_sh = gray_not.reshape((1,64))

print(gray_sh.shape)
plt.imshow(gray_sh,cmap='gray')
```

```python
test = (gray_sh/255.0)*15
test.shape
```
