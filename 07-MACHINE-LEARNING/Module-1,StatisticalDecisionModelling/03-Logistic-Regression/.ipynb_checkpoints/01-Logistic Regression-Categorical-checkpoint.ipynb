```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
```

```python
# Load dataset
data = pd.read_csv('./dataset/framingham.csv')
data.head()
```

```python
data['education'].value_counts()
```

```python
data['education'].fillna(value=1.0,inplace=True)
```

```python
data['education'].value_counts()
```

```python
edu = pd.get_dummies(data['education'],prefix='edu',drop_first=True)
```

```python
data_new = pd.concat((data,edu),axis=1)
```

```python
data_new.drop(columns='education',inplace=True)
```
