```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
```

```python
# Load dataset
data = pd.read_csv('./dataset/bank-full.csv',sep=';')
data.head()
```

# Convert Categorical data into numerical

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

### Consider Job column

```python
print('INFO: number of categories in job =',data['job'].nunique())
data['job'].value_counts()
```

# Method -1
* Step-1: Convert into factors (numerical)

```python
data_method1 = data.copy()
```

```python
data_method1['job'] = label.fit_transform(data_method1['job'])
data_method1.head()
```

```python
job = pd.get_dummies(data_method1['job'],prefix='job')
job.head()
```
