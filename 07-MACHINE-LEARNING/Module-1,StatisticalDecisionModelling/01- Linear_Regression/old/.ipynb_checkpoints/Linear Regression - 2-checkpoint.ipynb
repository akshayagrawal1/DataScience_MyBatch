```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

plt.style.available

```python
plt.style.use('seaborn')
```

**Loading data**

```python
data = pd.read_csv('data/Grade_Set_1.csv')
data.head()
```

```python
# Mean test grade 
data['Mean_grade'] = data.Test_Grade.mean()
data.head()
```

```python
fig=data.plot(x= 'Hours_Studied',y='Test_Grade',kind='Scatter',grid=True,s=100)
data.plot(x='Hours_Studied',y='Mean_grade',color='y',ax=fig)
plt.title('Hours Vs Grade')
plt.show()
```

```python
data['x - xbar']  = data['Hours_Studied'] - data['Hours_Studied'].mean()
data['y - ybar']  = data['Test_Grade'] - data['Test_Grade'].mean()
data['(x-xbar)(y-ybar)'] = data['y - ybar'] * data['x - xbar']
data['(x-xbar)sqr'] = data['x - xbar'] ** 2
data['(y-ybar)sqr'] = data['y - ybar'] ** 2
data.head()
```

```python
print('SSE @ mean= ',data['(y-ybar)sqr'].sum())
```

```python
data.corr()
```

```python
b1 = data['(x-xbar)(y-ybar)'].sum() / data['(x-xbar)sqr'].sum()
print('b1 = ',b1)
```

```python
b0  = data['Test_Grade'].mean() - b1 * data['Hours_Studied'].mean()
print('b0 = ',b0)
```

# $ y = b_0 + b_1 *x $ 

```python
y_pred = b0+ b1*data['Hours_Studied']
```

```python
data['yhat'] = y_pred
data
```

```python
data['(y -yhat)sqr'] = (data['Test_Grade'] - data['yhat'])**2
data
```

```python
print('new SSE @ yhat',data['(y -yhat)sqr'].sum())
```

```python
fig=data.plot(x= 'Hours_Studied',y='Test_Grade',kind='Scatter',grid=True,s=100)
data.plot(x='Hours_Studied',y='Mean_grade',color='y',ax=fig)
data.plot(x='Hours_Studied',y='yhat',color='r',ax=fig)
plt.title('Hours Vs Grade')
plt.show()
```

```python
# if student studied for 6.5 hours expected results
hr = 6.5
y_pred = b0+ b1*hr
print('Expected Grade if he studied {} hours is {}'.format(hr,y_pred))
```
