### Discriptive Stats

```python
#join plot
#sns.relplot('speed','dist',data=cars,kind='scatter')

sns.jointplot('Big Mac Price in USD','Hourly Wages in USD',data=cars,kind='scatter')
plt.show()
sns.jointplot('Big Mac Price in USD','Hourly Wages in USD',data=cars,kind='kde')
plt.show()
```

```python
import statsmodels.formula.api as smf
```

```python
cars.keys()
```

```python
data = cars.copy()
```

```python
data.columns = ['Country','BigMacPrice','HourlyWages']
```

```python
model = smf.ols(formula="HourlyWages~BigMacPrice",data=data)
results = model.fit()
```

```python
print(results.summary())
```

# Covariance

```python
cars.cov()
```

>### Correlation coefficient (*r* )

```python
cars.corr()
```

```python
results.params
```

```python
results.rsquared
```

```python
import statsmodels.api as sm
```

```python
fig = plt.figure(figsize=(12,8))
fig=sm.graphics.plot_regress_exog(results,'BigMacPrice',fig=fig)
```
