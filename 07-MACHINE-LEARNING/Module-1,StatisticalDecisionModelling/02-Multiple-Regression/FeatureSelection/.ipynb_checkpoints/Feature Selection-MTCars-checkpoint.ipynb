```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mtcars = pd.read_excel('./dataset/MTcars.xlsx')
mtcars.head()
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
model1 = smf.ols('mpg~hp+wt',data = mtcars).fit()
model1.summary()
```

```python
model2 = smf.ols('mpg~hp+wt+qsec',data = mtcars).fit()
model2.summary()
```

```python
model3 = smf.ols('mpg~cyl+disp+hp+drat+wt+qsec+vs+am+gear+carb',data = mtcars).fit()
model3.summary()
```
