```python
import pandas as pd
```

```python
with open('./lab/shakespeare-macbeth.txt') as f:
    text = f.read()
    f.close()
    
print(text)
```

```python
import re
```

```python
text = text.replace("don't","do not")
# lower case
text = text.lower()
text = re.sub(r'[^a-z]',' ',text)
text
```

```python
# Tokenization -> word tokens
from nltk import word_tokenize
```

```python
tokens = word_tokenize(text)
```

```python
# Bigrams
from nltk import trigrams
tri = list(trigrams(tokens))
print(tri)
```

```python
# Convert into Dataframe
df = pd.DataFrame(tri,columns=['1st','2nd','3rd'])
df.head(3)
```

```python
from collections import defaultdict
```

```python
model = defaultdict(lambda: defaultdict(lambda: 0))
```

```python
# convert three coloumns in two coloumns
for w1,w2,w3 in tri:
    model[(w1,w2)][w3] +=1
    #       key1   key2
```

```python
for w12 in model.keys():
    #print(w12,model[w12])
    # calculated count of key1
    total_count = sum(model[w12].values())
    #print(w12, 'total count =',total_count)
    for w in model[w12]:
        #print('{} count = {}'.format(w,model[w12][w]))
        # computing probability
        model[w12][w] /= total_count
```
