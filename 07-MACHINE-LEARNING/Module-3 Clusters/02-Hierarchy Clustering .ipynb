# Hierarchy Clustering 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**Load Dataset**

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset.head()
```

```python
dataset.info()
```

> **Extracting independent and dependent**

```python
X = dataset.iloc[:,[3,4]].values
```

> ** Data Visualisation**
- Scatter 

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],color='gray',marker='*',s=200)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

# Identifing Cluster
- Dendrograms

!pip install scipy

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(10,6))
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.xlabel('Cluster')
plt.ylabel('euclidean')
plt.title('Mall Custmers')
plt.show()
```

** From dendrogram no. of clusters = 4**

```python
n = 4 
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=n)
```

```python
y = hc.fit_predict(X) # training and identify my 'y'
```

** Visualisation**

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],marker='*',s=200,c=y)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Hierarchy Clustering')
plt.show()
```
