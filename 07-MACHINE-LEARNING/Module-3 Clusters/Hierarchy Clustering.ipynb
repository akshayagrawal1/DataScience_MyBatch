#  Hierarchy Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

**Load Dataset**

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset.head()
```

```python
dataset.info()
```

**Extracting independent variables**

```python
X = dataset.iloc[:,[3,4]].values
```

**Data Visualisation**
- scatter plot

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],color='gray',marker ='*',s = 200)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Score')
plt.title('Mall Customers')
plt.show()
```

# Identifying CLusters
- Dendrogram

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(20,8))
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.xlabel('No. of Clusters')
plt.ylabel('euclidean')
plt.title('Dendrogram')
plt.show()
```

**From above dendrogram , no. of cluster = 4**

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
n = 4
hc = AgglomerativeClustering(n_clusters=n)
```

```python
y = hc.fit_predict(X)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],marker ='*',s = 200,c=y)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Score')
plt.title('Mall Customers - Hierachy Clustering')
plt.show()
```
